### 1- Scheme

```python
dot = Digraph('scheme')

dot.node('R', 'Raw', style='filled', fillcolor='#DDA0DD')  # Light purple
dot.node('T', 'TPCData', style='filled', fillcolor='#ADD8E6')  # Light blue
dot.node('S', 'SiData, \nModularData', style='filled', fillcolor='#ADD8E6')  # Light blue
dot.node('G', 'MergerData', style='filled', fillcolor='#98FB98')  # Light green

dot.edge('R', 'T', label='Clusterize:\n3D points to tracks\nactroot -r tpc', labeldistance='-5.5', labelangle='-45')
dot.edge('R', 'S', label='Calibrate +\nthresholds:\nRaw to MeV\nactroot -r sil')
dot.edge('T', 'G', label='Filter:\nSplit, merge and\nimprove tracks\nactroot -f')
dot.edge('S', 'G', label='Points: vertex, silicon\nAngles\nEnergy loss in ACTAR\nTrack length\nactroot -m')

dot
```
